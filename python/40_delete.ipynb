# CRUD: Delete

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## Insert one book

We want to make sure this book is in the collection, so we can delete it

```python
platero =({
    "_id": "platero",
    "title": "platero y yo",
    "year": 1900
})

books.insert_one(platero)
```

### Find the book using the `_id` field

```python
plateroDoc = books.find_one({"_id": "platero"})

print(plateroDoc)
```

### Delete that book

```python
deleteResult = books.delete_one({"_id": "platero"})

print(deleteResult)
```
