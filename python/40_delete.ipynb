# CRUD: Delete

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
uri = "mongodb://admin:mongodb@localhost:27017/"

# Create a MongoDB Client
client = MongoClient(uri)
library_database = client["library"]

# Get the Books Collection
books = library_database["books"]
```

## Insert one book

We want to make sure this book is in the collection, so we can delete it

```python
books.insert_one({
    "_id": "platero",
    "title": "platero y yo",
    "year": 1900
})
```

### Find the book using the `_id` field

```python
books.find_one({"_id": "platero"})
```

### Delete that book

```python
deleted_book = books.delete_one({"_id": "platero"})

print(deleted_book.deleted_count)
```
