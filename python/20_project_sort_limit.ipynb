# CRUD: Project, Sort & Limit

## Startup code

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## Sort books by descending number of pages

```python
projection = {
    "title": 1,
    "pages": 1
}

cursor = books.find({}, projection).sort("pages", -1).limit(10)

for book in cursor:
    print(book)
```

## Limit

```python
projection = {
    "title": 1,
    "pages": 1
}

cursor = books.find(
    {}, # empty filter, all books
    projection
).sort("pages", -1).limit(10)

for book in cursor:
    print(f"{book['title']} - {book['pages']}")
```

## Chaining methods with find

```python
projection = {
    "title": 1,
    "pages": 1
}

cursor = books.find({}, projection).sort("pages", -1).skip(4).limit(5)

for book in cursor:
    print(f"{book['title']}")
```
