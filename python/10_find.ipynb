# CRUD: find

## Startup code

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## Find one book

```python
books.find_one()
```

## Find books written after 2008

```python
results = books.find({ "year": { "$gt" : 2008 }})

for f in results:
    print(f) 
```

## Exercise: find all the books that have less than 50 pages and print only the title and pages

```python
filter = { 
    "$and": [
        { "pages": { "$lt": 50 }}
    ]
}

projection = { "title": 1, "pages": 1 }

cursor = books.find(filter, projection).limit(10)

for book in cursor:
    print(book) 
```

## Find all the books that have less than 50 pages and project only the title and pages

```python
filter = { 
    "$and": [
        { "pages": { "$lt": 50 }}
    ]
}

projection = { "title": 1, "pages": 1 }

cursor = books.find(filter, projection).limit(10)

for book in cursor:
    print(book)
```

## Exclude `_id` field

```python
filter = { 
    "$and": [
        { "pages": { "$lt": 50 }},
    ]
}

projection = {
    "_id": 0,
    "title": 1,
    "pages": 1,
}

cursor = books.find(filter, projection).limit(10)

for f in cursor:
    print(f) 
```
