# Full Text Search - Create FTS Index

## Startup code

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## create_search_index()

https://www.mongodb.com/docs/languages/python/pymongo-driver/current/indexes/#create-a-search-index

```python
books.create_search_index = {
    "name": "default",
    "definition": {
        "mappings": {
            "dynamic": True
        }
    }
}
```

## Check that the index has been created

```python
# List Existing Search Indexes
indexes = list(books.list_indexes())
for index in indexes:
    print(f"Indexes: {indexes}")
```
