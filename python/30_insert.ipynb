```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
uri = "mongodb://admin:mongodb@localhost:27017/"

# Create a MongoDB Client
client = MongoClient(uri)
library_database = client["library"]

# Get the Books Collection
books = library_database["books"]
```

## Insert one book

```python
books.insert_one({
    "title": "El Quijote",
    "year": 1500
})
```

### Read the book we just inserted, using the year

```python
books_from_1500 = books.find({"year": 1500})

for book in books_from_1500:
    print(book)
```
