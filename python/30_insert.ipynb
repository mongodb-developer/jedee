# CRUD: Insert

## Startup code

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## Insert one book

```python
elQuijote = books.insert_one({
    "title": "El Quijote",
    "year": 1500
})

print(f'Inserted book with id: {elQuijote.inserted_id}')
```

### Read the book we just inserted, using the year

```python
books_from_1500 = books.find({"year": 1500})

for book in books_from_1500:
    print(book)
```
