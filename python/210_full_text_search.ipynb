# Full Text Search - $search

## Startup code

```python
# Import the MongoDB Driver
from pymongo import MongoClient

# Set your connection String
mongoDBURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true";

# Create a MongoDB Client
client = MongoClient(mongoDBURI)

# Get the Library Database
db = client["library"]

# Get the Books Collection
books = db["books"]
```

## $search

```python
pipeline = [
    {
        "$search": {
            "index": "default",
            "text": {
                "query": "Charity",
                "path": ["title", "authors.name", "genres"]
            }
        }
    }
]

results = books.aggregate(pipeline)

for doc in results:
    author_name = doc.get("authors", [{}])[0].get("name")
    print(f'Author: {author_name}')
```

## Fuzzy matching

```python
pipeline = [
    {
        "$search": {
            "index": "default",
            "text": {
                "query": "Chariti",
                "path": ["title", "authors.name", "genres"],
                "fuzzy": { "maxEdits": 2 }
            }
        }
    },
]

results = books.aggregate(pipeline)

for doc in results:
    print(f'Author: {doc}')
```
