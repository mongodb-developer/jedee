# Full Text Search - $search


## Startup code

```python
#Import the MongoDB Driver
from pymongo import MongoClient

# Connect to the MongoDB Server
mongoURI = "mongodb://admin:mongodb@localhost:27017/?directConnection=true"
client = MongoClient(mongoURI)

# Access a Database
db_name = "library"
db = client[db_name]

# Access a Collection
books = db["books"]

```

## $search

```python
pipeline = [
    {
        "$search": {
            "index": "default",
            "text": {
                "query": "Charity",
                "path": ["title", "authors.name", "genres"]
            }
        }
    }
]

results = books.aggregate(pipeline)

for book in results:
    author_name = book.get("authors", [{}])[0].get("name")
    print(f'Author: {author_name} Title: "{book["title"]}", Genres: {book.get("genres")}')
```

## Fuzzy matching

```python
pipeline = [
    {
        "$search": {
            "index": "default",
            "text": {
                "query": "Chariti",
                "path": ["title", "authors.name", "genres"],
                "fuzzy": { "maxEdits": 2 }
            }
        }
    },
]

results = books.aggregate(pipeline)

for book in results:
    author_name = book.get("authors", [{}])[0].get("name")
    print(f'Author: {author_name} Title: "{book["title"]}", Genres: {book.get("genres")}')
```
